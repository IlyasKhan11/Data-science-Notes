```python
df['day']=df['Date_of_Journey'].str.split('/').str[0]
df['month']=df['Date_of_Journey'].str.split('/').str[1]
df['year']=df['Date_of_Journey'].str.split('/').str[2]
df['day']=df['day'].astype(int)
df['month']=df['month'].astype(int)
df['year']=df['year'].astype(int)
df.drop('Date_of_Journey',axis=1,inplace=True)
df.head()
```

```python
# df['Arrival_Time']=df['Arrival_Time'].str.split(' ').str[0]
df['Arrival_Time']=df['Arrival_Time'].apply(lambda x:x.split(' ')[0])
df
```

```python
df['Arrival_hour']=df['Arrival_Time'].str.split(':').str[0]
df['Arrival_min']=df['Arrival_Time'].str.split(':').str[1]
df['Arrival_hour']=df['Arrival_hour'].astype(int)
df['Arrival_min']=df['Arrival_min'].astype(int)
df.drop('Arrival_Time',axis=1,inplace=True)
df.drop
df.head()
```

```python
df['Dep_hour']=df['Dep_Time'].str.split(':').str[0]
df['Dep_min']=df['Dep_Time'].str.split(':').str[1]
df['Dep_hour']=df['Dep_hour'].astype(int)
df['Dep_min']=df['Dep_min'].astype(int)
df.drop('Dep_Time',axis=1,inplace=True)
df.head()
```

```python
df['Total_Stops'].unique()
```

```python
modeTotalStops = df['Total_Stops'].mode() # Get most frequent value
print(modeTotalStops)
df['Total_Stops'].fillna(1, inplace=True)  # Fill NaN values with mode

# Convert categorical stops to numerical values
df['Total_Stops'] = df['Total_Stops'].replace({
    'non-stop': 0, 
    '1 stop': 1, 
    '2 stops': 2, 
    '3 stops': 3, 
    '4 stops': 4
})

# df
df.head()
```

```python
df.drop('Route',axis=1,inplace=True)

df.head()
```

```python
# Split the 'Duration' column into hours and minutes
df['Duration_hour'] = df['Duration'].apply(lambda x: int(x.split('h')[0]) if 'h' in x else 0)
df['Duration_min'] = df['Duration'].apply(lambda x: int(x.split('m')[0].split()[-1]) if 'm' in x else 0)

# Drop the original 'Duration' column
df.drop('Duration', axis=1, inplace=True)

df.head()
```

```python
df
```

```python
from sklearn.preprocessing import OneHotEncoder
encoder=OneHotEncoder()

# Apply OneHotEncoder to 'Airline', 'Source', and 'Destination'
encoded_features = encoder.fit_transform(df[['Airline', 'Source', 'Destination']]).toarray()

# Create a DataFrame for the encoded features
encoded_df = pd.DataFrame(encoded_features, columns=encoder.get_feature_names_out(['Airline', 'Source', 'Destination']))

# Concatenate the encoded features with the original DataFrame
df = pd.concat([df.reset_index(drop=True), encoded_df.reset_index(drop=True)], axis=1)

# Drop the original columns
df.drop(['Airline', 'Source', 'Destination'], axis=1, inplace=True)

df.head()
```
